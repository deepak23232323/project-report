#Visualising the Data

```python
empatrSrns.countplot(x='Attrition', data=EmpAtri, palette='viridis')
empatrplts.title('Distribution of Employee Attrition')
empatrplts.show()
```

```python
empatrplts.hist(EmpAtri['Age'], bins=20, color='skyblue', edgecolor='black')
empatrplts.title('Age Distribution of Employees')
empatrplts.xlabel('Age')
empatrplts.ylabel('Frequency')
empatrplts.show()
```

```python
empatrSrns.boxplot(x='Attrition', y='MonthlyIncome', data=EmpAtri, palette='coolwarm')
empatrplts.title('Monthly Income vs Attrition')
empatrplts.show()
```

```python
object_columns = EmpAtri.select_dtypes(include=['object']).columns

print("Object columns:", object_columns)
```

```python
Attrencd = empatrlncd()
EmpAtri['Attrition'] = Attrencd.fit_transform(EmpAtri['Attrition'])
EmpAtri['BusinessTravel'] = Attrencd.fit_transform(EmpAtri['BusinessTravel'])
EmpAtri['Department'] = Attrencd.fit_transform(EmpAtri['Department'])
EmpAtri['EducationField'] = Attrencd.fit_transform(EmpAtri['EducationField'])
EmpAtri['Gender'] = Attrencd.fit_transform(EmpAtri['Gender'])
EmpAtri['JobRole'] = Attrencd.fit_transform(EmpAtri['JobRole'])
EmpAtri['MaritalStatus'] = Attrencd.fit_transform(EmpAtri['MaritalStatus'])
EmpAtri['OverTime'] = Attrencd.fit_transform(EmpAtri['OverTime'])
```
